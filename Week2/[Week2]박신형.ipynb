### overfitting & underfitting

1. 오버피팅(overfitting)이란?  
\[답변] 지엽적인 특성까지 반영하여, high variance 하게 train되어, 새로운 데이터에 대해서는 예측하지 못하는 모델

2. 언더피팅(underfitting)이란?  
\[답변] 일부 특성만 반영하여, too bias하게 train되어, 새로운 데이터도 막 예측해버리는 모델

3. underfitting 막는 방법 2가지  
\[답변]1. parameter가 많은 모델을 사용한다<br/>
      2.더 좋은 feature을 제공하도록 한다

4. overfitting 판단 방법  
\[답변] train data에 대해서는 잘 예측되지만 test data에 대해서는 잘 예측되지 않는 것

5. overfitting 막는 방법  
\[답변] 모델을 단순화 하기, 특성 수를 줄이기

### 1. KNN(k-Nearest Neighbors) 

1. KNN 알고리즘 정리  
\[답변] K-최근접 이웃 알고리즘은 특정공간내에서 입력과 제일 근접한 k개의 요소를 찾아, 더 많이 일치하는 것으로 분류하는 알고리즘입니다.

* 참조 사이트 : https://woolulu.tistory.com/1
*  : https://doorbw.tistory.com/175

```python
# 아래 pass 부분을 지우고 내용을 채워주세요.
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
# iris data 로드
iris = load_iris()
# data split
x_train, x_test, y_train, y_test = train_test_split(iris.data, iris.target, train_size = 0.3, random_state = 0)

knn = KNeighborsClassifier(n_neighbors = 1, n_jobs = -1)
# n_neighbors 가 뜻하는 것? [답변] 가장 가까운 이웃을 몇개를 고를지에 대한 parameter
# n_jobs 가 뜻하는 것? [답변] 사용할 코어의 갯수 입니다. -1이면 모든 코어
knn.fit(x_train, y_train)      # x_train, y_train 이용해 트레이닝

prediction = knn.predict(x_test)       # knn 이용해 x_test 예측
results = [iris.target_names[p] for p in prediction]
print('results의 처음 6개 ==> {}'.format(results[:6]))

# 정확도 계산
print('score 메소드 이용 ==> {:.3f}'.format(knn.score(x_test, y_test)))         # x_test, y_test를 이용해 정확도 측정
```

(1) 위의 표 상황에서의 Entropy 구하기  
\[답변] E(WinterFamilyPhoto) =   - ( 1/4 )log( 1/4 ) - ( 3/4 )log( 3/4 ) = 0.24

(2) Cartoon 기준으로 나눴을 때, Information gain 구하기  <br/>
Entropy(0+, 2-) = - ( 2/2 )log( 2/2 ) = 0 <br/>
Entropy(1+, 1-) = - ( 1/2 )log( 1/2 ) - ( 1/2 )log( 1/2 ) = 0.30 <br/>

```python
# 아래 pass 부분을 지우고 내용을 채워주세요.
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.tree import export_text
# iris data 로드
iris = load_iris()
# data split
x_train, x_test, y_train, y_test = train_test_split(iris.data, iris.target, test_size = 0.3, random_state = 0)
# classifier 객체 생성
decision_tree = DecisionTreeClassifier(random_state = 0, max_depth = 2)
# max_depth 의미? [답변]  최대 깊이
decision_tree.fit(x_train, y_train)
r = export_text(decision_tree) 
print(r)

prediction = decision_tree.predict(x_test)    # decision tree 이용해 x_test 예측
results = [iris.target_names[p] for p in prediction]
print('results의 처음 6개 ==> {}'.format(results[:6]))

# 정확도 계산
print('score 메소드 이용 ==> {:.3f}'.format(decision_tree.score(x_test, y_test)))         # x_test, y_test를 이용해 정확도 측정
```
